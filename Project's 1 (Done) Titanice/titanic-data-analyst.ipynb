```python
# Identify missing values
missing_values = titanic.isnull().sum() 
missing_values# Fill missing values in the 'Age' column with the median age
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)

# Fill missing values in the 'Embarked' column with the most common value
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)

# Fill missing values in the 'Cabin' column with the most common value
titanic['Cabin'].fillna(titanic['Cabin'].mode()[0], inplace=True)
```

3-Next, the missing values in the `Age` column are filled with the median age using the `fillna()` function. The `fillna()` function replaces missing values with the specified value, in this case, the median age of the column.

4-Similarly, missing values in the `Embarked` column are filled with the most common value using the `fillna()` function and the `mode()` function. The `mode()` function returns the most frequently occurring value in a column.

5-Lastly, missing values in the `Cabin` column are filled with the most common value using the same approach as the `Embarked` column.
