# Import the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exploration Data Analysis (EDA)

# Load the dataset.

```python
# Display the first few rows of the dataset:-

data=pd.read_csv("E:/Data Set/New folder (9)/marketing_campaign.csv")
data.head(30)
```

# Check the dimensions of the dataset


```python
print(data.shape)
```

# Check the data types of each column

```python
print(data.dtypes)
```

# Check for missing values


```python
print(data.isnull().sum())
```

# Data Cleaning

```python
# Remove rows with missing values in the 'incom' column.
data = data.dropna(subset=['Income'])
```

```python
# Verify if missing values have been removed.
print(data.isnull().sum())
```

# Generate summary statistics of the dataset

```python
data.describe()
```

# Visualize the distribution of a numerical variable


```python
sns.histplot(data['Education'], kde=True)
plt.show()
```

```python
sns.histplot(data['Marital_Status'], kde=True)
plt.show()
```

# Visualize the relationship between two numerical variables


```python
sns.scatterplot(x='Education', y='Marital_Status', data=data)
plt.show()
```

# Explore relationships between variables:-

```python
correlation_matrix = data.corr() 
correlation_matrix
```

# Visualize the relationship between a numerical variable and a categorical variable

```python
sns.boxplot(x='Education', y='Income', data=data)
plt.show()
```

# Visualize the relationship between two categorical variables


```python
cross_tab = pd.crosstab(data['Marital_Status'], data['Education'])
cross_tab.plot(kind='bar', stacked=True)
plt.show()
```

# Visualize the data

```python
itemEducation = data['Education'].value_counts().index[:10]
itemValues = data['Education'].value_counts().values[:10]
plt.figure(figsize=(10,9))
plt.ylabel('count', fontsize='medium')
plt.xlabel('Education', fontsize='medium')
plt.title('Top 10 Education')
plt.bar(itemEducation,itemValues, width = 0.7,color='DeepPink',linewidth=0.4)
for i in range(len(itemEducation)):
 plt.text(i,itemValues[i],itemValues[i],ha='center',va='bottom')
plt.show()
```

```python
itemMarital_Status = data['Marital_Status'].value_counts().index[:10]
itemValues = data['Marital_Status'].value_counts().values[:10]
plt.figure(figsize=(10,9))
plt.ylabel('count', fontsize='medium')
plt.xlabel('Marital_Status', fontsize='medium')
plt.title('Top 10 Marital_Status')
plt.bar(itemMarital_Status,itemValues, width = 0.7,color='CornflowerBlue',linewidth=0.4)
for i in range(len(itemMarital_Status)):
 plt.text(i,itemValues[i],itemValues[i],ha='center',va='bottom')
plt.show()
```

```python
labels = data.Education.value_counts().index
colors = ['royalblue','orangered']
Education = data.Education.value_counts().values
plt.figure(figsize = (10,10))
plt.pie(Education, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('proportion of Education',color = 'black',fontsize = 20)
```

```python
plt.figure(figsize=(15,5))
plt.title('number of Education')
data.Education.value_counts().sort_index().plot()
```
